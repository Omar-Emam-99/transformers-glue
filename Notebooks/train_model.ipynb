```python
train_dataloader = DataLoader(data['train'], shuffle=True , batch_size=64)
eval_dataloader = DataLoader(data['dev'] , batch_size=64)

optimizer = AdamW(model.parameters(), lr=5e-5)

num_epochs = 1
num_training_steps = num_epochs * len(train_dataloader)

lr_scheduler = get_scheduler(name='linear' ,
                            optimizer=optimizer,
                            num_warmup_steps = 0,
                            num_training_steps=num_training_steps)
```

```python
from tqdm.auto import tqdm
import evaluate
from collections import defaultdict
```

```python
progress_bar = tqdm(range(num_training_steps))
model.to(device)
metric = evaluate.load("accuracy")
losses = defaultdict(list)
for epoch in range(num_epochs):
    model.train()
    for batch in train_dataloader:
        
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        
        loss = outputs.loss
        loss.backward()
        losses['train'].append(float(loss))
        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()
        
        progress_bar.update(1)
    
        model.eval()
    for batch in eval_dataloader:
        batch = {k: v.to(device) for k, v in batch.items()}
        with torch.no_grad():
            outputs = model(**batch)

        losses['eval'].append(float(outputs.loss))
        logits = outputs.logits
        predictions = torch.argmax(logits, dim=-1)
        #print(predictions)
        metric.add_batch(predictions=predictions, references=batch["labels"])
    
    print(metric.compute())
```

```python
eval_dataloader = DataLoader(data['test'] , batch_size=64)

for batch in eval_dataloader:
    batch = {k: v.to(device) for k, v in batch.items()}
    with torch.no_grad():
        outputs = model(**batch)

    logits = outputs.logits
    predictions = torch.argmax(logits, dim=-1)
    metric.add_batch(predictions=predictions, references=batch["labels"])

print(metric.compute())
```
